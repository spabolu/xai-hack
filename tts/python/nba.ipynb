```python
from nba_api. stats.endpoints import PlayByPlayV3
import pandas as pd

# Get play-by-play data
pbp = PlayByPlayV3(game_id="0022400092")
df = pbp.get_data_frames()[0]

# Select the most readable columns
readable_df = df[['period', 'clock', 'teamTricode', 'playerName', 'description', 'scoreHome', 'scoreAway']]

# Display nicely
pd.set_option('display.max_colwidth', None)  # Show full descriptions
pd.set_option('display.width', None)
print(readable_df. to_string())


from nba_api. live.nba. endpoints import PlayByPlay

pbp = PlayByPlay(game_id="0022400092")
data = pbp.get_dict()

# Access all action details
for action in data['game']['actions']:
    print(action)  # Contains all available fields
```
